# Missing Values Handle Karna

```python
df["Phone_Number"] = df["Phone_Number"].fillna("Not Available")
df["Last_Name"] = df["Last_Name"].fillna("Not Available")
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("Not Available")
```

# Phone Number Clean Karna
**Line 1: astype(str)**
- Phone Number ko string (text) bana deta hai 

```python
df["Phone_Number"] = df["Phone_Number"].astype(str)
```

**Line 2: str.replace**
- Text me se kuch hataana / replace karna
- Regex Pattern: r"[^\d]"
-  \d :- 
Means: digit (09)

-  [ ] :- 
Character group

-  ^ :-
Negation (NOT)

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"[^\d]", "", regex=True)
df
```

# Empty Phone Numbers Detect Karo

```python
df["Phone_Number"] == ""
```

# Empty Phone Numbers Fill Karna
-  Option 1: "Not Available" se fill karo (Most Common)

```python
df.loc[df["Phone_Number"] == "", "Phone_Number"] = "Not Available"
df
```

# Name Column Clean
** Line 1: .str.strip()**
- Left & right side ke spaces remove karta hai

```python
df["Address"] = df["Address"].str.strip()
df["First_Name"] = df["First_Name"].str.strip()
df["Last_Name"] = df["Last_Name"].str.strip()
df
```

**Line 2: .str.title()**
- Har word ka first letter capital
- Baaki letters lowercase

```python
df["Address"] = df["Address"].str.title()
df["First_Name"] = df["First_Name"].str.title()
df["Last_Name"] = df["Last_Name"].str.title()
df
```

# Problem Example
- Name column me values aisi hain:
- "/wdfd"
- "/suraj"
- "rahul/sharma"
- "/ amit "


```python
df["Last_Name"] = df["Last_Name"].str.replace("/", "", regex=False)
df
```

# Replace Name

```python
df["Paying Customer"] = df["Paying Customer"].replace({
    "Y": "Yes",
    "N": "No"
})
df
```

```python
df["Do_Not_Contact"] = df["Paying Customer"].replace({
    "Y": "Yes",
    "N": "No"
})
df
```
